1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
"""Функция, выделяющая из столбца "Образование и ВУЗ" информацию об уровне образования""" 
def education_fun(education):
    # Поскольку гарантировано, что информация об уровне образования в первых 2-х словах,
    # создаем список слов и проверяем на наличие ключевых слов два перых элемента списка
    education_list = education.split(' ')[0:2]
    # Устанавливаем уровень образования по ключевому слову
    if 'специальное' in education_list:
        ed_level = 'среднее специальное'
    elif 'Среднее' in education_list:
        ed_level = 'среднее'
    elif 'Неоконченное' in education_list:
        ed_level = 'неоконченное высшее'
    elif 'Высшее' in education_list:
        ed_level = 'высшее'
    # учитываем возможность ошибки
    else:
        ed_level = 'не определено'
    return ed_level 

    
# Создаем столбец 'Образование', содержащий информацию об уровне обзования, полученную с помощью функции education_fun
hh_df['Образование'] = hh_df['Образование и ВУЗ'].apply(education_fun)

# Проверяем состояние элементов столбца 'Образование' на предмет корректности замены (должно присутствовать 4 категории)
print(hh_df['Образование'].value_counts())

# Удаляем столбец «Образование и ВУЗ»
hh_df.drop(['Образование и ВУЗ'], axis=1,inplace=True)
```

2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
"""Функция, выделяющая из столбца "Пол, возраст" возраст и пол"""
# Аргументы genderage - значение в яцейке таблицы,
# parameter - принимает значение 'gender'- если выделяем пол, 'age' - если выделяем возраст
def genderage_fun(genderage, parameter):
    # создаем список слов
    genderage_list = genderage.split(' ,  ')
    # блок функции для параметра 'gender'
    if parameter == 'gender':
        # В зависимости от заданного пола передает значение 'М' или 'Ж'
        if genderage_list[0] == 'Мужчина':
            gend = 'M'
        elif genderage_list[0] == 'Женщина':
            gend = 'Ж'
        # учитываем возможность ошибки
        else:
            gend = 'не определено'
        return gend
    # блок функции для параметра 'age'
    elif parameter == 'age':
        # Выделяет значение возраста и возвращает его целым числом
        age_list = genderage_list[1].split(' ')
        ag = int(age_list[0])
        return ag

  
# Создаем столбец 'Пол', содержащий информацию о гендерной принадлежности, с помощью функции genderage_fun
hh_df['Пол'] = hh_df['Пол, возраст'].apply(lambda genderage: genderage_fun(genderage, 'gender'))
# Создаем столбец 'Возраст', содержащий информацию о возрасте, с помощью функции genderage_fun
hh_df['Возраст'] = hh_df['Пол, возраст'].apply(lambda genderage: genderage_fun(genderage, 'age'))
  
# Проверяем состояние элементов столбца 'Пол' на предмет корректности замены (должно присутствовать 2 категории)
print(hh_df['Пол'].value_counts(normalize=True))
print()

# Определяем средний возраст соискателя
print('Cредний возраст соискателей:', round(hh_df['Возраст'].mean(),1))

# Удаляем столбец «Пол, возраст»
hh_df.drop(['Пол, возраст'], axis=1,inplace=True)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
"""Функция, определяющая количество месяцев работы по данным столбца "Опыт работы"""
def get_experience(arg):
    # Если занчение не является строкой (вариант NaN), возвращаем NaN
    if type(arg) is not str: 
        return np.nan
    # Если значение является строкой, разделяем в список
    else:
        expirience_list = arg.split(' ')
        # исходя из строения предложения находим значение количества лет и месяцев, переводим в месяцы
        # если в указанной позиции стоит "мес", значит в значении указаны годы и месяцы
        if expirience_list[5][:3] == 'мес':
            # умножаем количество лет на 12 и прибавляем количество месяцев
            expitiens_mounth = int(expirience_list[2])*12 + int(expirience_list[4])
        # иначе указаны только годы или месяцы, определяем это по првой букве 4-го слова
        else:
            ex_end = expirience_list[3][0]
            # если это "г" или "л", указаны годы
            if ex_end == 'л' or ex_end == 'г':
                # умножаем число лет на 12
                expitiens_mounth = int(expirience_list[2])*12
            # если это "м", указаны только месяцы
            elif ex_end == 'м':
                # передаем число месяцев
                expitiens_mounth = int(expirience_list[2])
            else:
               # учитываем возможность ошибки
               expitiens_mounth = 'ошибка' 
        return expitiens_mounth


# Заменяем значения "не указано" на NaN
hh_df['Опыт работы'] = hh_df['Опыт работы'].apply(lambda x: np.nan if x == 'Не указано' else x)

# Создаем столбец "Опыт работы (месяц)", содержащий информацию об опыте работы в месяцах, с помощью функции get_experience
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_experience).round(0)

# Удаляем столбец «Опыт работы»
hh_df.drop(['Опыт работы'], axis=1,inplace=True)

# Находим медианный опыт работы
print('Медианный опыт работы:', hh_df['Опыт работы (месяц)'].median(), 'месяцев')
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

    Список городов-миллионников:

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
"""Функция, сортирующая города на четыре категории: «Москва», «Санкт-Петербург», «город-миллионник» и «другие»"""
def moving_function(arg):
    # разделяем слова в строке, создаем список слов
    moving_list = arg.split(' , ')
    # проверяем город проживания по первому слову в списке
    сity_residence = moving_list[0]
    if сity_residence == 'Москва':
        сity = 'Москва'
    elif сity_residence == 'Санкт-Петербург':
        сity = 'Санкт-Петербург'
    # если город относится к списку городов моллионников, заданному в списке million_cities, присваиваем значение 'город-миллионник'     
    elif сity_residence in million_cities:
        сity = 'город-миллионник'
    # для остальных городов присваиваем значение 'другие'
    else:
        сity = 'другие'
    return сity

"""Функция, определяющая готовность к переезду"""
def willing_relocate(arg):
    # разделяем слова в строке, создаем список слов
    moving_list = arg.split(' , ')
    # определяем элемент списка, содержащий информацию о готовности к переезду
    # если в списке присутствует информация о станции метро (первый элемент содержит "м."), то нам требуется 2 элемент
    if 'м.' in moving_list[1]:
        relocate = moving_list[2]
    # иначе (информации о станции метро нет) нам требуется первый элемент списка
    else:
        relocate = moving_list[1]
    # разделяем спроку по пробелу и проверяем первое слово
    relocate_list = relocate.split(' ')
    # если оно характеризует состояние как готовность к переежду, возвращаем True
    if relocate_list[0] == 'готов' or relocate_list[0] == 'готова' or relocate_list[0] == 'хочу':
        willing = True
    # иначе возвращаем False
    elif relocate_list[0] == 'не':
        willing = False
    # условие для проверки на случай ошибки
    else:
        willing = 'ошибка'
    return willing

"""Функция, определяющая готовность к командировкам"""
def willingness_travel(arg):
    # разделяем слова в строке, создаем список слов
    moving_list = arg.split(' , ')
    # задаем условия поиска элемента списка, соделжащего описание готовности к командирокам
    # если элементов 4 и есть элемент с информацией о метро, нам требуетя 3-й элемент
    if 'м.' in moving_list[1] and len(moving_list) == 4:
        travel = moving_list[3]
    # если элементов 3 и нет элемента с информацией о метро, нам требуетя 2-й элемент
    elif 'м.' not in moving_list[1] and len(moving_list) == 3:
        travel = moving_list[2]
    # иначе информация о готовности к командировкам пропущена и возвращаем занчение False
    else:
        return False
    # разделяем спроку по пробелу и проверяем первое слово
    travel_list = travel.split(' ')
    # если оно характеризует состояние как готовность к командировкам, возвращаем True
    if travel_list[0] == 'готов' or travel_list[0] == 'готова'or travel_list[0] == 'г':
        willing = True
    # иначе возвращаем False
    elif travel_list[0] == 'не':
        willing = False
    # условие для проверки на случай ошибки
    else:
        willing = 'ошибка'
    return willing

    
# Задаем список городов милионников
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']    
# Создаем столбец "Город", содержащий информацию о городе проживания с помощью функции moving_function  
hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(moving_function)

# Создаем столбец "Готовность к переезду" с помощью функции willing_relocate  
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(willing_relocate)

# Создаем столбец "Готовность к командировкам", содержащий информацию о городе проживания с помощью функции willingness_travel  
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(willingness_travel)


# Удаляем столбец «Город, переезд, командировки»
hh_df.drop(['Город, переезд, командировки'], axis=1,inplace=True)

# Выводим на печать распределение мест проживания соискателей
print(hh_df['Город'].value_counts(normalize=True))
print()

# Опредеяем процент соискателей, готовых и к командировкам и к переездам от общего числа соискателей
# Группируем тамбицу по признаку готовности к командировкам
hh_2 = hh_df.groupby(by='Готовность к командировкам').sum(numeric_only=True)
# Получаем из группировочной таблицы поличество людей со значением True по обеим признакам и делим на общее количество соискателей 
n = round(100*hh_2['Готовность к переезду'][True]/hh_df.shape[0])
print(f'Одновременно и к переездам, и к командировкам готовы {n}% соискателей.')

    
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
"""Функция, проверяющая наличие элемента в строке, если он присутствует, возвращает True, иначе False"""
def employment(empl, col):
    # создаем список из элементов строки
    employment_list = empl.split(', ')
    # если передаваемый элемент col имеется в строке, возвращаем True, иначе False
    if col in employment_list:
        return True
    else:
        return False

        
## Преобразуем категориальные признаки методом One Hot Encoding

# создаем список категорий желаемой занятости
employment_state = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
# проходим по элементам списка, проверяем наличие данных элементов в ячейках столбца 'Занятость'
# создаем новые столбцы соответствующие элементам списка, маркируем наличие проверяемых категорий с помощью функции employment
for col in employment_state:
    hh_df[col] = hh_df['Занятость'].apply(lambda empl: employment(empl, col))

# создаем список желаемых графиков работы
employment_state2 = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']
# проходим по элементам списка, проверяем наличие данных элементов в ячейках столбца 'График'
# создаем новые столбцы соответствующие элементам списка, маркируем наличие проверяемых категорий с помощью функции employment
for col in employment_state2:
    hh_df[col] = hh_df['График'].apply(lambda empl: employment(empl, col))
    
# Удаляем столбец «Занятость»
hh_df.drop(['Занятость'], axis=1,inplace=True)
# Удаляем столбец «График»
hh_df.drop(['График'], axis=1,inplace=True)

# отвечаем на контрольные вопросы
# определяем, сколько людей ищут проектную работу и волонтёрство (в обоих столбцах стоит True)
n = hh_df[hh_df['проектная работа'] == True]['волонтерство'].sum()
print(f'Проектную работу и волонтёрство одновременно в анкете указали {n} человек')
# определяем, cколько людей хотят работать вахтовым методом и с гибким графиком (в обоих столбцах стоит True)
m = hh_df[hh_df['вахтовый метод'] == True]['гибкий график'].sum()
print(f'Вахтовый метод и гибкий график одновременно в анкете указали {m} человек')
```

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Чтение данных о валютных курсах
rate_data = pd.read_csv('data/ExchangeRates.csv', sep=',')
# Копируем таблицу
rate_df = rate_data.copy()
# Преобразуем дату в формат datetime
rate_df['date'] = pd.to_datetime(rate_df['date'], format = '%d/%m/%y')
# Удаляем не нужные для анализа столбцы
rate_df.drop(['per', 'time', 'vol'], axis=1,inplace=True)


# Преобразуем в формат datetime дату в таблице о вакансиях HeadHunter
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], dayfirst=True).dt.date
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], dayfirst=True)

# Выделяем в отдельный столбец сумму желаемой зарплаты
hh_df['Сумма'] = hh_df['ЗП'].apply(lambda x: float(x.split(' ')[0]))
# Выделяем в отдельный столбец валюту, в которой указана зарплата
hh_df['Валюта'] = hh_df['ЗП'].apply(lambda x: x.split(' ')[1])
# Создаем словарь соотношения наименования валюты в таблице стандарту ISO                                      
currency_dict = {'грн.': 'UAH', 'USD': 'USD', 'EUR': 'EUR', 'бел.руб.': 'BYN', 'KGS': 'KGS', 'сум': 'UZS', 'AZN': 'AZN', 'KZT': 'KZT', 'руб.': 'RUB'}
# Заменяем наименования валюты в таблице на стандарт ISO
hh_df['Валюта'] = hh_df['Валюта'].apply(lambda currency: currency_dict.get(currency))


# Объединяем таблицы данных о соискателе и о валютных курсах
# При объединении в присоединяемой таблице для строк с зарплатой в рублях возникунт NaN, заменим их на '1' 
hh_df = hh_df.merge(
    rate_df,
    left_on = ['Валюта', 'Обновление резюме'],
    right_on = ['currency', 'date'],
    how = 'left'
)

# Заменяем значение close и proportion для рубля на 1
hh_df['close'] = hh_df['close'].fillna(1)
hh_df['proportion'] = hh_df['proportion'].fillna(1)

# Добавляем стобец ЗП (руб), умножая сумму зарплаты на курс валюты и разделив на пропорцию
hh_df['ЗП (руб)'] = hh_df['Сумма']*hh_df['close']/hh_df['proportion']

# Удаляем не нужные для анализа столбцы
hh_df.drop(['ЗП', 'Сумма', 'Валюта', 'currency', 'date', 'close', 'proportion'], axis=1,inplace=True)


# Отвечаем на контрольный вопрос: Чему равна желаемая медианная заработная плата соискателей (в рублях)?
n = round(hh_df['ЗП (руб)'].median()/1000)
print(f'Желаемая медианная заработная плата соискателей составляет {n} тыс. руб.')
```
